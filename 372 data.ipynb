```python
# data of sleeping
F1 = np.array([-0.6,0,0.1,0])
t1 = np.array([0,0.2083,0.3125,0.417]) ## 0, 5, 7.5, 10

plt.plot(t1, F1, 'ro', label="original data")

# curvefit
popt, pcov = curve_fit(func,t1,F1,[5.3,1,4.7,-0.764])
t = np.linspace(0,0.5,20)
plt.plot(t, func(t, *popt), label="Fitted Curve")
print(popt)

def f(x):
    return -(popt[0]*(x+popt[3])) - (popt[1]/(x+popt[3])) - popt[2]
from scipy.optimize import fmin
print(fmin(func=f,x0=np.array([0.3])))

def fc(x):
    return -0.100321

y_1 = np.ones(20)
Y_1 = y_1*(-0.100321)
plt.plot(t,Y_1, label="lowwer bound")
plt.show()

from shapely.geometry import Point, LineString
line_1 = LineString(np.column_stack((t,func(t, *popt))))
line_2 = LineString(np.column_stack((t,Y_1)))
intersection = line_1.intersection(line_2)
print(intersection)
print(popt*5/0.100321)
```
