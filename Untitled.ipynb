```python
%pylab inline
%config InlineBackend.figure_format = 'retina'
from ipywidgets import interact
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def func(x, a, b, c, k):
    return (a*(x+k)) + (b/(x+k)) + c
    
def funcs(x, a, b, c, k):
    return (a*(-x+k)) + (b/(-x+k)) + c
    
def funq(x,a,b,c,k):
    return -a**2 *(x+k)+b*(x+k)+ c
```
